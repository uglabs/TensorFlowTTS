```python
import os
import random
import shutil
import sys
from tqdm.notebook import trange, tqdm
import shutil
from pathlib import Path

kids_path = Path('../../../../data/kids_speech')
# libri_path = "../../../LibriTTS" # absolute path to TensorFlowTTS.
dataset_path = "../../../libritts" # Change to your paths. This is a output of re-format dataset.
subset = "kids_speech"

wavDir = kids_path / 'wav'
targetDir = kids_path / 'wavs'
import soundfile as sf
```

```python
wavs = [f.stem[:4] for f in wavDir.glob('*.wav')]
speakers = list(set(wavs))
for speaker in tqdm(speakers):
    targetSubDir = targetDir / speaker
    targetSubDir.mkdir(parents=True, exist_ok=True)
    for src_file in wavDir.glob('{}*.wav'.format(speaker)):
        dst_file = targetSubDir / src_file.name
        if dst_file.exists():
            continue
        shutil.copy(src_file, dst_file)

# dataset_path = targetDir
```

```python
def prepare(in_dir):
    # base_name = "kids_speech"
    metadata = Path(in_dir) / 'trans' / 'reps.txt'
    data = metadata.read_text().split('\n')
    i_start = data.index('--- BEGIN DATA ---') + 1
    print(i_start)
    speaker_map = {}
    #for i in range(i_start, len(data[i_start:], 3):
    for i in tqdm(range(i_start, len(data[i_start:]), 4)):
        row = data[i].split('  ')
        uttr_id = row[0]
        n_instsances = row[1]
        in_text = ' '.join(row[2:])
        text = in_text
        base_name = data[i+1].split(' ')[0]
        
        gender = base_name[0]
        assert base_name[-1] == '2', "{} is not two, in {}".format(base_name[-1], base_name)
        speaker = base_name[1:-1].replace(uttr_id, '')

        ID = gender + speaker
        # if config["preprocessing"]["speaker_id"] and config["preprocessing"]["speaker_id"] != speaker_id:
        #     continue
        wav_path = Path(in_dir) / 'wavs' / ID / '{}.wav'.format(base_name)
        # print(ID, wav_path, wav_path.exists())
        speaker_map[wav_path.name] = text
    return speaker_map

text_data = prepare(kids_path)
```

```python
# with open(os.path.join(libri_path, "SPEAKERS.txt")) as f:
# with open(kids_path / 'trans' / 'reps.txt')
    # data = f.readlines()

    
dataset_info = {}
max_speakers = 100  # Max number of speakers to train on
min_len = 5  # Min len of speaker narration time
max_file_len = 15  # max audio file length
min_file_len = 2  # min audio file length
```

```python
ids = os.listdir(targetDir)
possible_map = {}
# subset_path = os.path.join(libri_path, subset)
for i in os.listdir(targetDir):
    if i in ids:
        id_path = os.path.join(targetDir, i)
        id_dur = 0
        id_included = []
        
        for f_path in Path(id_path).glob("*.wav"):
            sf_file =  sf.SoundFile(str(f_path))
            dur = len(sf_file) / sf_file.samplerate
            if max_file_len < dur or dur < min_file_len:
                continue
            else:
                id_included.append(f_path)
                id_dur += dur
        possible_map[i] = {"dur": id_dur, "included": id_included}
```

```python
for speaker in tqdm(ids):
    # input
    srcSpeakerDir = Path(targetDir) / speaker
    # speakerDir.mkdir(parents=True, exist_ok=True)
    for wav in srcSpeakerDir.glob("*.wav"):
        if wav.name not in text_data:
            continue
        # print(text_data[wav.name])

        # outpu
        trgDirsrcSpeakerDir = Path(dataset_path) / speaker
        trgDirsrcSpeakerDir.mkdir(parents=True, exist_ok=True)
        # with open(trgDirsrcSpeakerDir / w)
        print(wav)
        shutil.copy(wav, trgDirsrcSpeakerDir / wav.name)
        txt_name = wav.with_suffix('.txt').name
        with open(trgDirsrcSpeakerDir / txt_name, 'w') as fd:
            fd.write(text_data[wav.name])
    # print(list(speakerDir.glob("*.wav")))
```
